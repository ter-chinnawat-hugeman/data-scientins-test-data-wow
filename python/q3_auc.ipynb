```python
import math
from typing import Callable

def integrate(function: Callable[[float], float], from_x: float, to_x: float) -> float:

    if from_x == to_x:
        return 0.0
    if from_x > to_x:
        return -integrate(function, to_x, from_x)

    steps = 10_000 # the bigger the n step , the smaller the interval . 

    h = (to_x - from_x) / steps

    total = 0.0

    total += 0.5 * function(from_x)
    total += 0.5 * function(to_x)

    x = from_x + h
    for _ in range(1, steps):
        total += function(x)
        x += h

    return total * h
```

```python
# y = x
assert abs(integrate(lambda x: x, 0, 1) - 0.5) < 1e-4

# y = x^2 
assert abs(integrate(lambda x: x*x, 0, 1) - (1/3)) < 1e-4

# y = sin(x)
assert abs(integrate(lambda x: math.sin(x), 0, math.pi) - 2.0) < 1e-4
```

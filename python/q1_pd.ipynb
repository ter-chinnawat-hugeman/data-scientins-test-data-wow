```python
# fillna() behavior :  fillna(value) to fill missing values in a series 
# with a specific value like coalescing operator in SQL
pd.Series([True, None]).fillna(True)

# 0    True
# 1    True
# dtype: bool
```

```python
# dropna() behavior : dropna() to drop rows with missing values in a series 
pd.Series([True, None]).dropna()

# 0    True
# dtype: object
```

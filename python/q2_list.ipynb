```python
# global variable list with length 1 
basket = [3]

def add_basket(item_id: int, basket: list=[])->list:
    # accept item_id and list as parameter 
    # default value of basket is empty list
    # empty list is evaluated only once, when the function is defined
    # return the pointer of the list
    basket.append(item_id)
    return basket

# call function without basket parameter, result in global basket shadowing by default empty list
# the global basket is not affected, since pointing to different memory location
# the function return the pointer of the list but no variable to assign to it
# default basket stored 120, now default basket become [120]
add_basket(120)

# default basket stored 5, now default basket become [120, 5]
add_basket(5)

# global basket is passed to function, so it is affected
# now the global basket is [3, 11]
add_basket(11, basket)

# call function without basket parameter, result in global basket shadowing by default empty list
# the global basket is not affected, since pointing to different memory location
# default basket stored 10
# now default basket become [120, 5], 10
print(add_basket(10))

# print the global basket, it is [3, 11]
print(basket)

# To be honest , the global basket is was I expected. 
# But the default basket is not. Without running the code, I thought the default basket will be [10]. Clearly I was wrong


# the output is
# [120, 5, 10]
# [3, 11]
```
